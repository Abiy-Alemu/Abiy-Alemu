1. Longest Increasing Subsequence (LIS)
Problem:
Given an unsorted array of integers, find the length of the longest increasing subsequence. A subsequence is defined as a sequence derived by deleting some or no elements from the array without changing the order of the remaining elements.
Example:
arr = [10, 22, 9, 33, 21, 50, 41, 60, 80]
# The longest increasing subsequence is [10, 22, 33, 50, 60, 80], so the output should be 6.

Hint:
Use dynamic programming or binary search for optimization. The brute force solution has a time complexity of O(2^n), but you can optimize it to O(n^2) using dynamic programming, or even O(n log n) with binary search and a clever approach.


```python
from typing import List

# class Solution:
def lengthOfLIS(arr: List[int]) -> int:
    n = len(arr)
    dp = [1]*n

    for i  in range(1,n):
        for j in range(i):
            if arr[i]> arr[j]:
                dp[i] = max(dp[i],dp[j]+1)

    return max(dp)
#Example:
arr = [10, 22, 9, 33, 21, 50, 41, 60, 80]
result = lengthOfLIS(arr)
print(result)
```

#2. Sudoku Solver
Problem:
Write a function that takes a partially filled 9x9 Sudoku board and fills in the missing values to solve it. If the board is unsolvable, return False.
#Example:
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]
`The function should fill in the board and return True if solved.`
Hint:
This problem can be solved using `backtracking`. Try to fill each empty space with numbers from 1 to 9, and recursively check if the number placement is valid. If the board is valid, continue; if not, backtrack and try a different number.

```python
def solveSudoku(board: list[list[int]]) -> bool:
    """
    Solves a Sudoku puzzle by filling in the empty cells.
    Modifies the board in-place.
    """
    for r in range(9):
        for c in range(9):
            if board[r][c] == 0:  # Find an empty cell
                for num in range(1, 10):  # Try numbers 1 to 9
                    if isValid(board, r, c, num):
                        board[r][c] = num
                        if solveSudoku(board):  # Recurse
                            return True
                        else:
                            board[r][c] = 0  # Backtrack
                return False  # No valid number found for this cell
    return True  # All cells filled, puzzle solved

def isValid(board: list[list[int]], r: int, c: int, num: int) -> bool:
    """
    Checks if placing 'num' at (r, c) is valid according to Sudoku rules.
    """
    # Check row
    for x in range(9):
        if board[r][x] == num:
            return False

    # Check column
    for x in range(9):
        if board[x][c] == num:
            return False

    # Check 3x3 subgrid
    start_row = r - r % 3
    start_col = c - c % 3
    for i in range(3):
        for j in range(3):
            if board[i + start_row][j + start_col] == num:
                return False
    return True

# Example usage:
board = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

print("Original board:")
for row in board:
    print(row)

if solveSudoku(board):
    print("\nSolved board:")
    for row in board:
        print(row)
else:
    print("\nNo solution exists.")
```

#3. N-Queens Problem
Problem:
Solve the N-Queens problem, which asks you to place N queens on an NxN chessboard such that no two queens threaten each other. A queen can attack another queen if they are on the same row, column, or diagonal.
Example:
For N = 4, one valid solution would be:
[ [0, 1, 2, 3],
  [1, 3, 0, 2],
  [2, 0, 3, 1],
  [3, 2, 1, 0] ]

Hint:
You can solve this problem using backtracking. Keep track of columns, main diagonals, and anti-diagonals where queens have already been placed to avoid conflicts.


```python
def solveNQueens(n: int) -> list[list[int]]:
    solutions = []

    # board_state[r] = c means a queen is at (r, c)
    # Initialize with -1 to indicate no queen is placed yet in that row
    board_state = [-1] * n

    # Sets to keep track of occupied columns and diagonals
    cols = set()         # Columns already occupied by a queen
    pos_diag = set()     # Positive diagonals (r + c) occupied by a queen
    neg_diag = set()     # Negative diagonals (r - c) occupied by a queen

    def backtrack(r):
        # Base case: if all rows have queens, a valid solution is found
        if r == n:
            solutions.append(list(board_state)) # Add a copy of the current board state
            return

        # Try placing a queen in each column of the current row 'r'
        for c in range(n):
            # Check for conflicts
            if c in cols or (r + c) in pos_diag or (r - c) in neg_diag:
                # Conflict found, try the next column
                continue

            # If no conflict, place a queen
            cols.add(c)
            pos_diag.add(r + c)
            neg_diag.add(r - c)
            board_state[r] = c

            # Recurse to the next row
            backtrack(r + 1)

            # Backtrack: remove the queen to explore other possibilities
            cols.remove(c)
            pos_diag.remove(r + c)
            neg_diag.remove(r - c)
            board_state[r] = -1 # Optional: reset for clarity

    backtrack(0) # Start the backtracking process from the first row (row 0)
    return solutions

# Example usage for N = 4:
n_val = 4
results = solveNQueens(n_val)
print(f"\nFor N = {n_val}, there are {len(results)} distinct solutions:")
for i, sol in enumerate(results):
    print(f"Solution {i+1}: {sol}")

# Example usage for N = 8 (to show the number of solutions):
n_val_8 = 8
results_8 = solveNQueens(n_val_8)
print(f"\nFor N = {n_val_8}, there are {len(results_8)} distinct solutions.")
# Uncomment the following lines to print all solutions for N=8 if desired:
# for i, sol in enumerate(results_8):
#     print(f"Solution {i+1}: {sol}")
```

4. Word Ladder
Problem:
Given two words (beginWord and endWord), and a dictionary of words, find the shortest transformation sequence from beginWord to endWord, such that:
Only one letter can be changed at a time.
Each transformed word must exist in the dictionary.
For example:
beginWord = "hit"
endWord = "cog"
wordList = ["hot", "dot", "dog", "lot", "log", "cog"]
# The shortest transformation sequence is: "hit" -> "hot" -> "dot" -> "dog" -> "cog"

Hint:
Use breadth-first search (BFS) to explore the word transformation graph. Each word is a node, and each letter change forms an edge. Keep track of the visited nodes to avoid cycles and redundant work.

```python
from collections import deque

def ladderLength(beginWord: str, endWord: str, wordList: list[str]) -> int:
    word_set = set(wordList)
    if endWord not in word_set:
        return 0

    queue = deque([(beginWord, 1)]) # (word, length)
    visited = {beginWord}

    while queue:
        current_word, current_length = queue.popleft()

        if current_word == endWord:
            return current_length

        # Find all possible next words by changing one letter
        for i in range(len(current_word)):
            original_char = current_word[i]
            for char_code in range(ord('a'), ord('z') + 1):
                new_char = chr(char_code)
                if new_char == original_char:
                    continue

                next_word_list = list(current_word)
                next_word_list[i] = new_char
                next_word = "".join(next_word_list)

                if next_word in word_set and next_word not in visited:
                    visited.add(next_word)
                    queue.append((next_word, current_length + 1))

    return 0 # No transformation sequence found

# Example usage:
beginWord = "hit"
endWord = "cog"
wordList = ["hot", "dot", "dog", "lot", "log", "cog"]

result = ladderLength(beginWord, endWord, wordList)
print(f"The shortest transformation sequence length from '{beginWord}' to '{endWord}' is: {result}")

beginWord2 = "hhit"
endWord2 = "ccg"
wordList2 = ["hot", "dot", "dog", "lot", "log", "cog"]

result2 = ladderLength(beginWord2, endWord2, wordList2)
print(f"The shortest transformation sequence length from '{beginWord2}' to '{endWord2}' is: {result2}")
```

#5. Find the Median of Two Sorted Arrays
Problem:
Given two sorted arrays nums1 and nums2, find the median of the two sorted arrays. The overall run time complexity should be O(log(min(n, m))) where n and m are the lengths of the arrays.
Example:
nums1 = [1, 3]
nums2 = [2]
# The median is 2.0

Hint:
This is a classic binary search problem. Instead of merging the arrays (which takes O(n+m) time), use binary search to partition the arrays in such a way that the left part and the right part of the merged array are balanced.


```python
from typing import List

def findMedianSortedArrays(nums1: List[int], nums2: List[int]) -> float:
    # Ensure nums1 is the shorter array for easier binary search logic
    if len(nums1) > len(nums2):
        nums1, nums2 = nums2, nums1

    n1 = len(nums1)
    n2 = len(nums2)

    # Perform binary search on the shorter array (nums1)
    low = 0
    high = n1

    while low <= high:
        partitionX = (low + high) // 2  # Partition point for nums1
        # partitionY ensures total elements in left half is (n1 + n2 + 1) // 2
        partitionY = (n1 + n2 + 1) // 2 - partitionX

        # Determine elements at the boundaries of partitions
        # If partition is 0, then there are no elements on the left, use -infinity
        # If partition is length of array, then no elements on the right, use +infinity
        maxX = -float('inf') if partitionX == 0 else nums1[partitionX - 1]
        minX = float('inf') if partitionX == n1 else nums1[partitionX]

        maxY = -float('inf') if partitionY == 0 else nums2[partitionY - 1]
        minY = float('inf') if partitionY == n2 else nums2[partitionY]

        if maxX <= minY and maxY <= minX:
            # We found the correct partition
            if (n1 + n2) % 2 == 0:  # Even number of total elements
                return (max(maxX, maxY) + min(minX, minY)) / 2
            else:  # Odd number of total elements
                return float(max(maxX, maxY))
        elif maxX > minY:
            # partitionX is too far right, move left
            high = partitionX - 1
        else:
            # maxY > minX, partitionX is too far left, move right
            low = partitionX + 1

    # Should not reach here for valid inputs, but for completeness
    return 0.0

# Example usage:
nums1_ex1 = [1, 3]
nums2_ex1 = [2]
median_ex1 = findMedianSortedArrays(nums1_ex1, nums2_ex1)
print(f"Median of {nums1_ex1} and {nums2_ex1} is: {median_ex1}") # Expected: 2.0

nums1_ex2 = [1, 2]
nums2_ex2 = [3, 4]
median_ex2 = findMedianSortedArrays(nums1_ex2, nums2_ex2)
print(f"Median of {nums1_ex2} and {nums2_ex2} is: {median_ex2}") # Expected: 2.5

nums1_ex3 = [0, 0]
nums2_ex3 = [0, 0]
median_ex3 = findMedianSortedArrays(nums1_ex3, nums2_ex3)
print(f"Median of {nums1_ex3} and {nums2_ex3} is: {median_ex3}") # Expected: 0.0

nums1_ex4 = []
nums2_ex4 = [1]
median_ex4 = findMedianSortedArrays(nums1_ex4, nums2_ex4)
print(f"Median of {nums1_ex4} and {nums2_ex4} is: {median_ex4}") # Expected: 1.0

nums1_ex5 = [2]
nums2_ex5 = []
median_ex5 = findMedianSortedArrays(nums1_ex5, nums2_ex5)
print(f"Median of {nums1_ex5} and {nums2_ex5} is: {median_ex5}") # Expected: 2.0
```

#6. Graph Cycle Detection (Directed Graph)
Problem:
Given a directed graph, determine if there is a cycle in the graph. If there is a cycle, return True; otherwise, return False.
Example:
graph = {
    0: [1],
    1: [2],
    2: [3],
    3: [0]
}
# This graph contains a cycle, so the result should be True

Hint:
This is a classic graph traversal problem. Use Depth First Search (DFS) with a recursion stack (or a visited list) to track the state of each node. If you revisit a node that is already in the recursion stack, you have detected a cycle.


```python
from collections import defaultdict

def hasCycle(graph: dict) -> bool:
    visited = set()  # Stores all nodes that have been visited
    recursion_stack = set()  # Stores nodes in the current DFS path

    def dfs(node):
        visited.add(node)
        recursion_stack.add(node)

        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                if dfs(neighbor):
                    return True
            elif neighbor in recursion_stack:
                # Cycle detected: neighbor is already in the current DFS path
                return True

        recursion_stack.remove(node)  # Backtrack: remove node from current path
        return False

    # Iterate through all nodes to handle disconnected components
    for node in graph:
        if node not in visited:
            if dfs(node):
                return True
    return False

# Example usage:
graph1 = {
    0: [1],
    1: [2],
    2: [3],
    3: [0]  # Cycle: 0 -> 1 -> 2 -> 3 -> 0
}
print(f"Graph 1 has a cycle: {hasCycle(graph1)}") # Expected: True

graph2 = {
    0: [1, 2],
    1: [3],
    2: [3],
    3: []
}
print(f"Graph 2 has a cycle: {hasCycle(graph2)}") # Expected: False

graph3 = {
    0: [1],
    1: [2],
    2: [0]  # Cycle: 0 -> 1 -> 2 -> 0
}
print(f"Graph 3 has a cycle: {hasCycle(graph3)}") # Expected: True

graph4 = {
    0: [],
    1: [],
    2: []
}
print(f"Graph 4 has a cycle: {hasCycle(graph4)}") # Expected: False
```
